```python
import netCDF4 as ncdf
import pylab as P
import matplotlib.cm as cm
import glob
import numpy as np

field = "REFL_10CM"

dirs = ["/scratch/wof/realtime/20170518/EXPS/EI4/DART_OUTPUT/preassim_priorinf_mean.nc.20170518*"]
files = glob.glob(dirs[0])
files = sorted(files, key=lambda name: int(name[-4:]))

infl_mean = []
infl_max  = []
time = []

for file in files:
    ncdata = (ncdf.Dataset(file)).variables[field][0,0:35]
    time.append(np.int(file[-4:]))
    infl_mean.append( ncdata.mean() )
    infl_max.append( ncdata.max() )

print("Read in EI4")

plot_mean = {'MHN_EI_time': time, 'MHN_EI': infl_mean}
plot_max  = {'MHN_EI_time': time, 'MHN_EI': infl_max}

#######################

dirs = ["/scratch/wof/realtime/20170518/EXPS/MHN/DART_OUTPUT/preassim_priorinf_mean.nc.20170518*"]
files = glob.glob(dirs[0])
files = sorted(files, key=lambda name: int(name[-4:]))

infl_mean = []
infl_max  = []
time = []

for file in files:
    ncdata = (ncdf.Dataset(file)).variables[field][0,0:35]
    time.append(np.int(file[-4:]))
    infl_mean.append( ncdata.mean() )
    infl_max.append( ncdata.max() )

    
print time

plot_mean['MHN_AI_time'], plot_mean['MHN_AI'] = time, infl_mean
plot_max ['MHN_AI_time'], plot_max['MHN_AI'] = time, infl_max

#######################
dirs = ["/scratch/wof/realtime/20170516/EXPS/MHN_EI6/DART_OUTPUT/preassim_priorinf_mean.nc.20170516*"]
files = glob.glob(dirs[0])
files = sorted(files, key=lambda name: int(name[-4:]))

infl_mean = []
infl_max  = []
time = []

for file in files:
    ncdata = (ncdf.Dataset(file)).variables[field][0,0:35]
    time.append(np.int(file[-4:]))
    infl_mean.append( ncdata.mean() )
    infl_max.append( ncdata.max() )

    
print time
             
plot_mean['MHN_EI6_time'], plot_mean['MHN_EI6'] = time, infl_mean
plot_max ['MHN_EI6_time'], plot_max['MHN_EI6'] = time, infl_max

#######################
dirs = ["/scratch/wof/realtime/20170516/EXPS/LSD/DART_OUTPUT/preassim_priorinf_mean.nc.20170516*"]
files = glob.glob(dirs[0])
files = sorted(files, key=lambda name: int(name[-4:]))

infl_mean = []
infl_max  = []
time = []

for file in files:
    ncdata = (ncdf.Dataset(file)).variables[field][0,0:35]
    time.append(np.int(file[-4:]))
    infl_mean.append( ncdata.mean() )
    infl_max.append( ncdata.max() )

    
print time
             
plot_mean['MHN_LSD_time'], plot_mean['MHN_LSD'] = time, infl_mean
plot_max ['MHN_LSD_time'], plot_max['MHN_LSD'] = time, infl_max


#######################
dirs = ["/scratch/wof/realtime/20170516/EXPS/VSD/DART_OUTPUT/preassim_priorinf_mean.nc.20170516*"]
files = glob.glob(dirs[0])
files = sorted(files, key=lambda name: int(name[-4:]))

infl_mean = []
infl_max  = []
time = []

for file in files:
    ncdata = (ncdf.Dataset(file)).variables[field][0,0:35]
    time.append(np.int(file[-4:]))
    infl_mean.append( ncdata.mean() )
    infl_max.append( ncdata.max() )

plot_mean['MHN_VSD_time'], plot_mean['MHN_VSD'] = time, infl_mean
plot_max ['MHN_VSD_time'], plot_max['MHN_VSD'] = time, infl_max

print("Read in MHN_VSD")

######################
dirs = ["/scratch/wof/realtime/20170516/EXPS/VSD/DART_OUTPUT/preassim_priorinf_sd.nc.20170516*"]
files = glob.glob(dirs[0])
files = sorted(files, key=lambda name: int(name[-4:]))

infl_mean = []
infl_max  = []
time = []

for file in files:
    ncdata = (ncdf.Dataset(file)).variables[field][0,0:35]
    time.append(np.int(file[-4:]))
    infl_mean.append( ncdata.mean() )
    infl_max.append( ncdata.max() )

plot_mean['VSD_time'], plot_mean['VSD'] = time, infl_mean
plot_max ['VSD_time'], plot_max['VSD'] = time, infl_max

print("Read in VSD")
```

```python
fig = P.figure(figsize = (12,10))


P.plot(plot_mean['MHN_AI_time'], plot_mean['MHN_AI'], color='r')
P.plot(plot_mean['MHN_EI_time'], plot_mean['MHN_EI'], color='g')
#P.plot(plot_mean['MHN_EI6_time'], plot_mean['MHN_EI6'], color='b')
P.plot(plot_mean['MHN_LSD_time'], plot_mean['MHN_LSD'], color='k')
P.plot(plot_mean['MHN_VSD_time'], plot_mean['MHN_VSD'], color='b')
P.xlabel = 'Time (UTC)'
P.ylabel = "Value of Inflation"
leg = P.legend(loc='best', ncol=2, mode="expand", shadow=True, fancybox=True)
P.tight_layout()
P.grid()
P.legend(('A2009 Adaptive Inflation', 'EI-default', 'EI-0.4SD', "EI-Variable SD"),
           loc='upper left', shadow=True)
P.title('Mean Inflation value in 3D domain')


P.show()
```
